```python
#  linear kernel
model = SVR(kernel="linear")
model.fit(X_train,y_train_scaled)
y_pred_scaled = model.predict(X_test)
print("r2_score for test:",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_score for train:",r2_score(y_train_scaled,y_train_pred_scaled))
```
