```python
import numpy as np
# from sklearn.linear_model import LinearRegression
```

```python
#  Linear Regression
class LinearRegression:
    def __init__(self,learning_rate=0.01,n_iter=1000):
        self.bias = None
        self.weights = None
        self.lr = learning_rate
        self.n_iter = n_iter
        


    def  fit(self,X,y):
         
         m,n =  X.shape
         #  numbers of rows and samples 
         #  step 1
         self.bias = 0
         self.weights = np.zeros(n)
        
        

         #  gradient descent
        
         for i  in range(self.n_iter):
              #  step 2 cal y_pred
             
             y_pred = self.bias + np.dot(X,self.weights) 
           #  feartures ke accordu=ing zeros values ayegi
         #  step3 gradient descent nikalana 
             db = (1/m) +np.sum(y_pred-y)
             dw = (1/m) + np.dot(X.T,(y_pred-y))
         #  step 4 convergence theorem parms update
             self.bias -= self.lr * db
             self.weights -= self.lr * dw
              
      
    def predict(self,X):
          y_pred  = self.bias + np.dot(X,self.weights)
          return y_pred
             
 
         


```

```python
X =  np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

print(y_pred)
print(model.bias)
print(model.weights)
```

```python
#  linear regression with ols
class LinearRegressionOLS:
    def __init__(self):
        self.bias = None
        self.weights = None
        

    def  fit(self,X,y):
         m,n = X.shape
         X_b =  np.c_[np.ones((m,1)),X]
        #  normal equation
        # theta = np.dot(np.linalg.inv(X_b.T,X_b),np.dot(X_b.T,y))
         theta = np.linalg.inv(X_b.T@X_b)@ X_b.T@ y
         self.bias = theta[0]
         self.weights = theta[1:]
       
 

       



  
    def predict(self,X):
          y_pred  = self.bias + np.dot(X,self.weights)
          return y_pred
             
```

```python
X =  np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])
model = LinearRegressionOLS()
model.fit(X,y)
y_pred = model.predict(X)

print(y_pred)
print(model.bias)
print(model.weights)
```
